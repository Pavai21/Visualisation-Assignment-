```python
x=np.arange(1,11)
y=3*x+5
plt.title("Line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,"--g*",linewidth=2,markersize=10)
```

```python
x=[1,2,3,4,5,6,7]
y=[160,150,140,145,175,165,180]
plt.plot(x,y, linewidth = 2)
plt.xlabel("x-axis", fontsize=13)
plt.ylabel("y-axis", fontsize=13)
plt.title("x vs y", fontsize=16)
```

```python
days = [1,2,3,4,5,6,7]
sales_1=[160,150,140,145,175,165,180]
sales_2=[70,90,160,150,140,145,175]
plt.figure(figsize=(8,4), dpi=150)
plt.plot(days, sales_1)
plt.plot(days, sales_2)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Days vs Sales')
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,"--r+",markersize=3)
plt.xlabel('x-axis',fontsize=13)
plt.ylabel('y-axis',fontsize=13)
plt.title('Line Plot', fontsize = 18)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin =np.sin(x)
y_cos =np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin,"r")
plt.subplot(2,2,1)
plt.plot(x,y_cos,"g")
plt.subplot(2,2,3)
plt.plot(x,y_sin,"r")
plt.subplot(2,2,4)
plt.plot(x,y_cos,"g")
```
